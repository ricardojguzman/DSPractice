```python
import matplotlib.pyplot as  plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x** 2
```

```python
x
```

```python
y
```

```python
# Functional method

plt.plot(x,y)
plt.xlabel(' X label')
plt.ylabel('Y label')
plt.title('Title')
#plt.show for no jupyter notebook
```

```python
#Colors
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1,0.2,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
# Plot inside 
fig= plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Bigger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#Different plots, using axes as array
fig,  axes = plt.subplots(nrows =1 , ncols=2)
#axes.plot(x,y)
#plt.tight_layout()

#for current_ax in axes:
 #   current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First Plot")



axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))


axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,label ='X Y')
ax.plot(y,x,label =' Y X')

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax.legend(loc=0)
#ax.legend(loc=(0.1,0.1))
```

```python
#Colors

fig = plt.figure()

ax=fig.add_axes([0,0,1,1])


#linewidth #linestyle #marker
ax.plot(x,y,color='orange',lw=3,alpha=1,ls='-',marker='o',markersize=10, markerfacecolor='blue', markeredgewidth=3,markeredgecolor='green') #or RGB color
```

```python
#set xli,
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
